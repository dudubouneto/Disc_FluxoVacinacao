```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Discretização das variáveis/Data/classes_finais/classes_finais_v3.csv')
# remover drs, ra, etc
data.drop(['codigo_ibge', 'nome_ra', 'nome_drs'], axis=1, inplace=True)
data.head()
```

```python
# transformando em "lista"
cities = list(data.iloc[index].astype(str).to_list() for index in range(0, len(data)))
```

```python
type(cities)
```

```python
cities[0]
```

```python
# transformando em uma tabela "one-hot"
from mlxtend.preprocessing import TransactionEncoder
a = TransactionEncoder()
a_data = a.fit(cities).transform(cities)
df = pd.DataFrame(a_data, columns=a.columns_)
df
```

```python
df = apriori(df, min_support=0.02, use_colnames=True)
df
```

```python
df_ar = association_rules(df, metric='confidence', min_threshold=0.6)
df_ar
```

```python
df_ar['antecedents'] = df_ar['antecedents'].apply(lambda x: ', '.join(list(x))).astype("unicode")
df_ar['consequents'] = df_ar['consequents'].apply(lambda x: ', '.join(list(x))).astype("unicode")
```

```python
df_ar['consequents'].value_counts()
```

```python
df_ar = df_ar.sort_values(by='lift', ascending=False)
```

```python
df_ar.to_csv('/content/drive/MyDrive/Colab Notebooks/Regras de Associaçao/assoc_rules_v2.csv', index=False)
```
